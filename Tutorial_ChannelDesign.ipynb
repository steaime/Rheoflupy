### 1: Constitutive equation

Goal: find the channel shape, $L(x)$, such that the flow of the continuous phase applies an oscillatory stress, $\sigma=\bar\sigma \sin(\omega t)$ to a flowing droplet

Constitutive equation for continuous phase: $\sigma = \eta \dot\varepsilon = \eta (\partial_x v_x +\partial_y v_y) / 2 = \eta \partial_x v_x$

Planar flow rate: $q=\int v_x(y)dy = \beta v_x(y=0) L(x)$, with $\beta$ a numeric prefactor that depends on the channel aspect ratio

Assuming $q/\beta \approx \textrm{const}$, we relate $\sigma$ to $L(x)$: 

$$
\sigma=-\frac{\eta q}{\beta L^2}\frac{dL}{dx}
$$

From which we get a constitutive equation for channel shape:

$$
\frac{dL}{dx} = \frac{L^2 \beta}{q\eta} \sigma(x)
$$

We want to prescribe a *time-dependent* stress, in the reference framework of a droplet flowing at speed $v_d = \zeta v_x = \zeta q/\beta L$. In that reference frame, time is:

$$
t(x) = \int\limits_0^x \frac{dx}{v_x} = \frac{\beta}{\zeta q} \int\limits_0^x Ldx
$$

Using this in the expression of $\sigma$, we get to the final constitutive equation:

$$
\frac{dL}{dx} = L^2 \frac{\bar\sigma\beta}{q\eta} \sin\left[ \frac{\omega \beta}{\zeta q} \int\limits_0^x Ldx \right]
$$

We express $L(x)$ in units of the constriction size, $L_0$: $\tilde L = L/L_0$

$$
\frac{d \tilde L}{dx} = \tilde L^2 \frac{L_0 \bar\sigma\beta}{q\eta} \sin\left[ \frac{\omega \beta L_0}{\zeta q} \int\limits_0^x \tilde Ldx \right]
$$

The quantity $k=\omega\beta L_0/\zeta q$ plays the role of a wavevector: we use it to define a nondimensional position: $\tilde x = kx = x \omega\beta L_0/ \zeta q$. This simplifies the expression:

$$
\frac{d \tilde L}{d\tilde x} = \tilde L^2 \frac{\bar\sigma \zeta}{\eta \omega} \sin\left( \int\limits_0^{\tilde x} \tilde Ld\tilde x \right) = \tilde L^2 \tilde \sigma \sin \left( \int\limits_0^{\tilde x} \tilde Ld\tilde x \right)
$$

The shape of the channel is dictated by just one dimensionless parameter, setting the stress amplitude: $\tilde \sigma = \bar\sigma \zeta/\eta \omega$. The solution, $\tilde L(\tilde x)$, is then expressed in physical units, scaling by $L_0$ along $y$ and by $1/k$ along $x$, to obtain the desired frequency at a given flow rate.

### 2: Numerical solution

To solve this equation numerically, we replace variable: $\tilde x \rightarrow \tilde t = \int\limits_0^{\tilde x} \tilde L d\tilde x = \omega\beta/\zeta q \int\limits_0^x L dx=\omega t$; $d \tilde t/d\tilde x = \tilde L$, which turns the integro-differential equation into a simple ODE:

$$
\frac{d \tilde L}{d\tilde t} = \frac{d \tilde L}{d\tilde x}\frac{d \tilde x}{d\tilde t} = \tilde L \tilde \sigma \sin \left(  \tilde t \right)
$$

After solving for $\tilde L(\tilde t)$ with initial condition $\tilde L(\tilde t=0)=1$, we compute $\tilde x=\int\limits_0^{\tilde t} d\tilde t/\tilde L$, and we express the result in a parametric form $(\tilde x(\tilde t), \tilde L(\tilde t))$.

Next, we express it in physical units by taking $L(\tilde t) = L_0 \tilde L(\tilde t)$ and $x(\tilde t) = \tilde x(\tilde t)/k$.  
Final solution will be: 

$$
(x, L) = \left(\frac{q \zeta}{\omega\beta L_0}\tilde x(\tilde t), L_0\tilde L(\tilde t)\right)
$$

### 4: Prescribe channel length

To draw microfluidic channels, it is useful to impose the total length of the channel. This is easiest in a stress sweep at constant $\omega$ and $q$ (where $k$ is also constant): in this case the shape of the channel can be computed in dimensionless units, and everything can be rescaled at once. Otherwise, we ask for the relative $k$ for each constriction.

To analyze the channel shape:

### 5: AutoCAD script

````python
def gen_script(fname, omega, sigma, channel_length, L0=1e-4, nperiods=2, npts=300, eta=1e-3, beta=1, startp=(0,0), plot=True):
    x, L, t, pars = cd.channel_set_length(omega=omega, sigma=sigma, channel_length=channel_length, L0=L0, eta=eta, 
                                      nperiods=nperiods, pts_per_sol=npts, beta=beta, return_params=True)
    hdr = '; DESIGN PARAMS: \n; L0 = {0:.1f} microns'.format(L0*1e6)
    if hasattr(omega, '__len__'):
        hdr += '\n; {0} frequencies: omega ranges from {1:.1f} to {2:.1f} [rad/s]'.format(len(omega), 
                                                                        np.min(omega), np.max(omega))
    else:
        hdr += '\n; fixed omega = {0:.1f} [rad/s]'.format(omega)
    if hasattr(sigma, '__len__'):
        hdr += '\n; {0} stress amplitudes: sigma ranges from {1:.3f} to {2:.3f} [Pa]'.format(len(sigma), 
                                                                        np.min(sigma), np.max(sigma))
    else: 
        hdr += '\n; fixed sigma = {0:.3f} [Pa]'.format(sigma)
    print(hdr)
    hdr += '\n; Common parameters:'
    hdr += '\n; - x0 = {0}'.format(startp)
    hdr += '\n; - L0 = {0:.1f} [um]'.format(pars[0]['L0']*1e6)
    hdr += '\n; - eta = {0:.1f} [mPa.s]'.format(pars[0]['eta']*1e3)
    hdr += '\n; - beta = {0:.3f}'.format(pars[0]['beta'])
    hdr += '\n; - zeta = {0:.3f}'.format(pars[0]['zeta'])
    for i in range(len(pars)):
        hdr += '\n; Constriction {0}: {1} periods, {2} points'.format(i+1, pars[i]['nperiods'], pars[i]['pts'])
        hdr += '\n; - sigma tilde = {0:.3f}'.format(pars[i]['sigma_tilde'])
        hdr += '\n; - k = {0:.2f} [1/mm]'.format(pars[i]['k']*1e3)
        hdr += '\n; - sigma = {0:.3f} [Pa]'.format(pars[i]['sigma_tilde']*pars[i]['eta']*pars[i]['omega']/pars[i]['zeta'])
        hdr += '\n; - omega = {0:.1f} [rad/s]'.format(pars[i]['omega'])
        hdr += '\n; - computed q = {0:.1e} [mm2/s]'.format(1e6*cd.q_from_k(omega=pars[i]['omega'], beta=pars[i]['beta'], 
                                                                           L0=pars[i]['L0'], k=pars[i]['k'], zeta=pars[i]['zeta']))
    with open(fname, 'w') as fout:
        fout.write('line')
        for i in range(len(t)):
            fout.write('\n{0:02d},{1:02d}'.format(int(L[i]/2*1e6)+startp[0], startp[1]-int(x[i]*1e6)))
        fout.write('\n\nline')
        for i in range(len(t)):
            fout.write('\n{0:02d},{1:02d}'.format(int(-L[i]/2*1e6)+startp[0], startp[1]-int(x[i]*1e6)))
        fout.write('\n\n' + hdr)
    if plot:
        csa.analyze_sweep(x, L, t, q=pars[0]['q'], eta=pars[0]['eta'])
````

````python
gen_script('out\channel_shape_1.scr', omega=100, sigma=np.geomspace(0.1, 1, 10), channel_length=0.01, L0=4e-5, eta=1e-2, nperiods=1, startp=(0,0))
````


### 6: Creep and square waves

For a step stress, $\sigma(t)=\sigma_0\Theta(t)$, with $\Theta$ Heaviside's Theta function, the channel shape has an analytical solution, which we find by solving:

$$
\frac{dL}{dx} = -\frac{L^2 \beta \sigma_0}{q\eta} \Theta\left[\frac{\beta}{\zeta q} \int\limits_0^x Ldx\right]=
    \begin{cases}
      0, & \text{if}\ x<0 \\
      -L^2 \beta\sigma_0 / q\eta, & \text{if}\ x>0
    \end{cases}
$$

This can be integrated by separation of variables:

$$
\int L^{-2}dL=\frac{1}{L_0}-\frac{1}{L}=-\frac{\beta \sigma_0}{q\eta}x
$$

$$
L(x) = \frac{L_0}{1+\lambda x}
$$

with $\lambda=\beta \sigma_0 L_0/q\eta$. In dimensionless units, $\tilde{L}=L/L_0$ and $\tilde{x}=x/L_0$:

$$
\tilde{L}(\tilde{x}) = \frac{1}{1+\tilde{\sigma} \tilde{x}}
$$

where $\tilde{\sigma}=\beta\sigma_0 L_0^2/q\eta$.


For $\sigma_0<0$ (diverging channel), there is a maximum channel length, $\lambda^{-1}=q\eta/\beta \sigma_0 L_0$, at which $L\rightarrow \infty$.

In general, a finite channel length $x_{max}$ is associated with a creep time

$$
t_{max} = \frac{\beta L_0}{\zeta q} \int\limits_0^{x_{max}} \frac{dx}{1+\lambda x} = \frac{\beta L_0}{\zeta q} \frac{\ln(1+\lambda x_{max})}{\lambda} = \frac{\eta}{\zeta \sigma_0}\ln\left(1+\frac{\beta\sigma_0L_0x_{max}}{q\eta}\right)
$$

Conversely, a prescribed creep time $t_{max}$ will require a channel length

$$
x_{max} = \frac{q\eta}{\beta\sigma_0L_0}\left[\exp\left(\frac{\zeta \sigma_0 t_{max}}{\eta}\right)-1\right]
$$

In dimensionless units: $\tilde{x}_{max} = \tilde{\sigma}^{-1}[\exp(\tilde{t}_{max})-1]$, with $\tilde{t}_{max}=\zeta\sigma_0 t_{max}/\eta$.

````python
fig, ax = plt.subplots()
for st in np.geomspace(1e-2, 100, 5):
    ax.plot(*cd.creep_dimensionless(st, 3), label=r'$\tilde{\sigma}=$' + str(st))
    ax.plot(*cd.creep_dimensionless(-st, -3), label=r'$\tilde{\sigma}=-$' + str(st))
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'$\tilde{x}$')
ax.set_ylabel(r'$\tilde{L}$')
ax.legend()
````

### 7: Startup and triangular waves

A stress ramp, $\sigma(t)=\sigma_0(t/\tau)$, also has an analytical solution, which we find by solving:

$$
\frac{dL}{dx} = -\frac{L^2 \beta \sigma_0}{q\eta \tau} \frac{\beta}{\zeta q} \int\limits_0^x Ldx
= -\frac{L^2 \beta^2 \sigma_0}{q^2\eta \tau \zeta}\int\limits_0^x Ldx
$$

In nondimensional units, $\tilde{L}=L/L_0$ and $\tilde{x}=x/L_0$:

$$
\frac{d\tilde{L}}{d\tilde{x}} = -\frac{L_0^4 \beta^2 \sigma_0}{q^2\eta \tau \zeta} \tilde{L}^2\int\limits_0^{\tilde{x}} \tilde{L}d\tilde{x} = -\tilde{\sigma} \tilde{L}^2\int\limits_0^{\tilde{x}} \tilde{L}d\tilde{x}
$$

with $\tilde{\sigma}=L_0^4 \beta^2 \sigma_0/q^2\eta \tau \zeta$.
We rewrite it as a simple ODE for $y(\tilde{x})=\int\tilde{L}d\tilde{x}$:

$$
\frac{d^2 y}{d\tilde{x}^2} = -\tilde{\sigma} \left(\frac{dy}{d\tilde{x}}\right)^2y(\tilde{x})
$$

We solve it by observing that:

$$
\frac{d^2 y}{d\tilde{x}^2} = \frac{d\tilde{L}}{d\tilde{x}} = \frac{d\tilde{L}}{dy} \frac{dy}{d\tilde{x}} = \tilde{L}\frac{d\tilde{L}}{dy} = -\tilde{\sigma}\tilde{L}^2y
$$

Therefore, we get to a simple ODE which we solve by separating variables:

$$
\frac{d\tilde{L}}{\tilde{L}} = -\tilde{\sigma}ydy
$$

$$
\tilde{L} = \frac{dy}{d\tilde{x}} = C\exp\left(-\frac{\tilde{\sigma}y^2}{2}\right)
$$

$$
\exp\left(\frac{\tilde{\sigma}y^2}{2}\right)dy = Cd\tilde{x}
$$

If $\tilde{\sigma}=-|\tilde{\sigma}|<0$, we have the integral of a Gaussian, which is an Error function:

$$
\int\limits_0^y\exp\left(-\frac{|\tilde{\sigma}|y^2}{2}\right)dy = \sqrt{\frac{\pi}{2|\tilde{\sigma}|}}\mathrm{erf}\left(y\sqrt{\frac{|\tilde{\sigma}|}{2}}\right)=C\tilde{x}+C_1
$$

$$
y=\sqrt{\frac{2}{|\tilde{\sigma}|}}\mathrm{erf}^{-1}\left[\sqrt{\frac{2|\tilde{\sigma}|}{\pi}}\left(C\tilde{x}+C_1\right)\right]
$$

$$
\tilde{L}=\frac{dy}{d\tilde{x}}=\sqrt{\frac{2}{|\tilde{\sigma}|}}\frac{d}{d\tilde{x}}\mathrm{erf}^{-1}\left[\sqrt{\frac{2|\tilde{\sigma}|}{\pi}}\left(C\tilde{x}+C_1\right)\right]
$$

We use that:

$$
\frac{d}{d\tilde{x}}\mathrm{erf}^{-1}\left[f(\tilde{x})\right]=\frac{\sqrt{\pi}}{2}\frac{df}{d\tilde{x}}\exp\left\{\mathrm{erf}^{-1}\left[f(\tilde{x})\right]\right\}^2
$$

We obtain:

$$
\tilde{L}(\tilde{x})=C \exp\left\{\mathrm{erf}^{-1}\left[\sqrt{\frac{2|\tilde{\sigma}|}{\pi}}\left(C\tilde{x}+C_1\right)\right]\right\}^2
$$

Applying boundary conditions:

$$
\tilde{L}(\tilde{x})=\exp\left\{\left[\mathrm{erf}^{-1}\left(\sqrt{\frac{2|\tilde{\sigma}|}{\pi}}\tilde{x}\right)\right]^2\right\}
$$

For small $\tilde{x}$, $\tilde{L}$ grows exponentially: $\tilde{L}(\tilde{x}\ll 1)\sim\exp\left(|\tilde{\sigma}|\tilde{x}^2\right)$

At large $\tilde{x}$, $\tilde{L}$ diverges as $\tilde{x}\rightarrow \sqrt{\pi/2|\tilde{\sigma}|}$

In terms of dimensional quantities (remember: $\tilde{\sigma}=L_0^4 \beta^2 \sigma_0/q^2\eta \tau \zeta$):

$$
L(x)=L_0 \exp\left\{\left[\mathrm{erf}^{-1}\left(\sqrt{\frac{2|\sigma_0|}{\pi\eta\tau\zeta}}\frac{L_0\beta}{q}x\right)\right]^2\right\}
$$

If $\tilde{\sigma}=|\tilde{\sigma}|>0$, the same calculations apply by replacing the error function ($\mathrm{erf}$) with the imaginary error function ($\mathrm{erfi}$). To recap:

$$
\tilde{L}(\tilde{x})=
    \begin{cases}
      \exp\left[\left(\mathrm{erf}^{-1}\left(\sqrt{-\frac{2\tilde{\sigma}}{\pi}}\tilde{x}\right)\right)^2\right], & \text{if}\ \tilde{\sigma}<0 \\
      \exp\left[-\left(\mathrm{erfi}^{-1}\left(\sqrt{\frac{2\tilde{\sigma}}{\pi}}\tilde{x}\right)\right)^2\right], & \text{if}\ \tilde{\sigma}>0 \\
    \end{cases}
$$

$$
L(x)=
    \begin{cases}
      L_0 \exp\left[\left(\mathrm{erf}^{-1}\left(\sqrt{\frac{-2\sigma_0}{\pi\eta\tau\zeta}}\frac{L_0\beta}{q}x\right)\right)^2\right], & \text{if}\ \sigma_0<0 \\
      L_0 \exp\left[-\left(\mathrm{erfi}^{-1}\left(\sqrt{\frac{2\sigma_0}{\pi\eta\tau\zeta}}\frac{L_0\beta}{q}x\right)\right)^2\right], & \text{if}\ \sigma_0>0 \\
    \end{cases}
$$

To relate the channel length to the time, we compute $L'=dL/dx$ at the end of the channel, to then compute the stress $\sigma_{max}=-q\eta L'_{max}/L^2=\sigma_0(t/\tau)$. Duration of channel will then be $t=\tau \sigma_{max}/\sigma_0=-q\eta \tau L'_{max}/\sigma_0 L^2$

$$
\frac{\tilde{L}'}{\tilde{L}^2}=\sqrt{2\tilde{\sigma}}\mathrm{erf}^{-1}\left(\sqrt{\frac{2\tilde{\sigma}}{\pi}}\tilde{x}\right)
$$

$$
t_{max}=-\frac{q\eta\tau}{\sigma_0}\frac{L'}{L^2}=-\frac{q\eta\tau}{\sigma_0L_0^2}\frac{\tilde{L}'}{\tilde{L}^2}=
-\frac{q\eta\tau}{\sigma_0L_0^2}\sqrt{2\tilde{\sigma}}\mathrm{erf}^{-1}\left(\sqrt{\frac{2\tilde{\sigma}}{\pi}}\tilde{x}\right)
$$

### 8: Generalized waveform

More complex stress profiles don't have an analytical solution, and need to be integrated from the equation:

$$
\frac{dL}{dx} = \frac{L^2 \beta}{q\eta} \sigma(t(x))
$$

with 

$$
t(x) = \int\limits_0^x \frac{dx}{v_x} = \frac{\beta}{\zeta q} \int\limits_0^x Ldx
$$
